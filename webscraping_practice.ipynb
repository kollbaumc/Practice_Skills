```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://bigten-online.com/big-ten-football-win-loss-records.html"

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
table1 = soup.find('table')
```

```python
print(table1)
```

```python
football_titles = table1.find_all('th')[0:10]
```

```python
football_titles
```

```python
football_table_titles = [title.text for title in football_titles]

print(football_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns =  football_table_titles)
```

```python
df
```

```python
columndata = table1.find_all('tr')
```

```python
for row in columndata [1:15]:
    rowdata = row.find_all('td')
    Individualdata = [data.text for data in rowdata]
    
    length = len(df)
    df.loc[length] = Individualdata
    
    
      
```
