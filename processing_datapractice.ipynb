```python
#importing pandas
import pandas as pd
```

```python
#reading the csv and only reading specific columns, this data is housing prices from zillow
HD = pd.read_csv('ziphousingdata.csv', usecols = ['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'State', 'City', 'Metro', 'CountyName', '7/31/2023', '8/31/2023'])
```

```python
#Here are our columns headers
HD.columns
```

```python
#only including rows where the state is Iowa
option = ['IA']
new_HD = HD[HD['State'].isin(option)]
new_HD
```

```python
#only including the rows where the metro area is Des Moines
option2 = ['Des Moines-West Des Moines, IA']
new_HD2 = HD[HD['Metro'].isin(option2)]
new_HD2
```

```python
#only including the rows where the city is ankeny
option3 = ['Ankeny']
new_HD3 = HD[HD['City'].isin(option3)]
new_HD3
```

```python
#checking for null values for the data includes rows where the state is Iowa
new_HD.isnull().any()
```

```python
#dropping any possible duplicates, making sure even though I don't believe there are any.  
new_HD = new_HD.drop_duplicates()
new_HD
```

```python
#Checking the data types of my columns
new_HD.dtypes
```

```python
#Changing the Metro column to a string type
new_HD['Metro'] = new_HD['Metro'].astype('str')
```

```python
#Making nan in metro column into a blank space
new_HD['Metro'] = new_HD['Metro'].str.replace('nan','')
```

```python
new_HD
```

```python
#Taking out all NaN values
new_HD =  new_HD.fillna('')
```

```python
#Getting rid of certain rows, zips in Linn County for instance.  I just picked something randomly.
for x in new_HD.index:
    if new_HD.loc[x, "CountyName"] == 'Linn County':
        new_HD.drop(x, inplace=True)
```

```python
#resetting the index
new_HD.reset_index(drop=True)
```
